```python
from bs4 import BeautifulSoup

# Load the HTML from a file (or you can use a string)
with open(r'C:\Users\ebowen\Box\Wu Lab\Project - statin\8. RNA-seq\Elizabeth\LW15 analysis\LW15-Functional-Screening\index.html', 'r', encoding='utf-8') as f:
    html_content = f.read()

# Parse the HTML
soup = BeautifulSoup(html_content, 'html.parser')

# Find the table with id 'functionalScreening'
table = soup.find('table', id='functionalScreening')

gene_ids = []
# Iterate over all rows in the table's tbody
for row in table.tbody.find_all('tr'):
    cells = row.find_all('td')
    # Check if there are enough cells (the gene id is in the 5th cell)
    if len(cells) >= 5:
        # Get the text from the 5th cell and strip any extra whitespace
        gene_id = cells[4].get_text(strip=True)
        gene_ids.append(gene_id)

# Print the list of gene IDs
print(gene_ids)
```

```python
from bs4 import BeautifulSoup

# Load the HTML file
html_path = r"c:\Users\ebowen\Box\Wu Lab\Project - statin\8. RNA-seq\Elizabeth\LW15 analysis\LW15-Functional-Screening\variance_genes.html"
with open(html_path, "r", encoding="utf-8") as file:
    soup = BeautifulSoup(file, "html.parser")

# Find the table with id "functionalScreening"
table = soup.find("table", {"id": "functionalScreening"})
if not table:
    print("Error: Table with id 'functionalScreening' not found.")
    exit()

# Find the "Found In" column header and its index
headers = table.find("thead").find_all("th")
found_in_index = None
for i, header in enumerate(headers):
    if header.get_text(strip=True) == "Found in":
        found_in_index = i
        header.decompose()  # Remove the header
        print(f"Removed 'Found In' column header at index {i}.")
        break

if found_in_index is None:
    print("Error: 'Found In' column not found.")
    exit()

# Remove the corresponding <td> elements in each row
rows = table.find("tbody").find_all("tr")
for row_index, row in enumerate(rows, start=1):
    cells = row.find_all("td")
    if len(cells) > found_in_index:
        del cells[found_in_index]  # Remove the cell at the "Found In" column index
        print(f"Row {row_index}: Removed 'Found In' column cell.")

# Save the modified HTML back to the file
with open(html_path, "w", encoding="utf-8") as file:
    file.write(str(soup))

print("Removed 'Found In' column from the table.")
```

```python
from bs4 import BeautifulSoup

# Load the HTML file
html_path = r"c:\Users\ebowen\Box\Wu Lab\Project - statin\8. RNA-seq\Elizabeth\LW15 analysis\LW15-Functional-Screening\variance_genes.html"
with open(html_path, "r", encoding="utf-8") as file:
    soup = BeautifulSoup(file, "html.parser")

# Find the table with id "functionalScreening"
table = soup.find("table", {"id": "functionalScreening"})
if not table:
    print("Error: Table with id 'functionalScreening' not found.")
    exit()

# Count the number of <th> headers to determine the correct number of columns
headers = table.find("thead").find_all("th")
num_columns = len(headers)
print(f"Number of columns based on headers: {num_columns}")

# Iterate through the rows in the table body
rows = table.find("tbody").find_all("tr")
for row_index, row in enumerate(rows, start=1):
    cells = row.find_all("td")
    if len(cells) > num_columns:
        # Remove extra <td> cells beyond the number of headers
        for extra_cell_index in range(num_columns, len(cells)):
            cells[extra_cell_index].decompose()
            print(f"Row {row_index}: Removed extra <td> cell at index {extra_cell_index}.")

# Save the modified HTML back to the file
with open(html_path, "w", encoding="utf-8") as file:
    file.write(str(soup))

print("Removed extra <td> cells from all rows.")
```
